```python
import re
import urllib.request
import matplotlib.pyplot as plt
from collections import Counter


url = 'https://en.wikipedia.org/wiki/Albert_Einstein'
response = urllib.request.urlopen(url)
text = response.read().decode('utf-8')
years = re.findall(r'\b(18[0-9]{2}|19[0-9]{2}|20[0-9]{2})\b', text)

year_counts = Counter(years)

sorted_years = sorted(year_counts.keys())
sorted_counts = [year_counts[year] for year in sorted_years]
print(sorted_counts)

plt.figure(figsize=(10, 6))
plt.hist(years, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Year Range (Binned)')
plt.ylabel('Frequency')
plt.title('Frequency of Years Mentioned in Albert Einstein\'s Wikipedia Article')
plt.tight_layout()
plt.show()
```
